```python
import re
import requests
import zipfile
import os
import numpy as np
import pandas as pd

with open('masterfilelist-translation.txt') as f:
    lines = f.readlines()
```

```python
re_1 = re.compile('.*gdeltv2/2022.*')
re_2 = re.compile('http.*')
re_3 = re.compile('.*(07)0000.translation.gkg.*')

list_1 = [re_2.search(x).group() for x in list(filter(re_1.match, lines))]
list_2 = list(filter(re_3.match, list_1))
```

```python
list_2
```

```python


for url in list_2:
    r = requests.get(url, allow_redirects=True)
    open('CSV files/zipfile.zip', 'wb').write(r.content) # download the zipfil>

    if zipfile.is_zipfile('CSV files/zipfile.zip'):
        with zipfile.ZipFile('CSV files/zipfile.zip', 'r') as zip_file:
            zip_file.extractall('CSV files') # Extract the content of the zipf>
    os.remove('CSV files/zipfile.zip') # remove the zipfile
```
